# 7. Plot the avg tip by size  (lineplot)

```python
# Answer
```

```python
tips.groupby('size').tip.mean().plot()
```

```python
sns.lineplot(data=tips, x="size", y="tip", ci= None)
```

# ci = int in [0, 100] or None, optional


Size of the confidence interval for the regression estimate. This will be drawn using translucent bands around the regression line. The confidence interval is estimated using a bootstrap; for large datasets, it may be advisable to avoid that computation by setting this parameter to None.

```python
sns.lineplot(data=tips, x="size", y="tip", ci= 80)
```

```python
df1= tips.groupby('size')['tip'].mean()
df1
```

```python
fig, ax = plt.subplots()
ax.plot(df1.index, df1.values)
ax.set_xlabel("size");
```

# 10. Set an additional dimension using size column

```python
#Answer
```

```python
tips.plot.scatter("tip", "total_bill", c="size", cmap="coolwarm")
```

```python
tips.plot.scatter("tip", "total_bill", c="size", cmap="Accent")
```

```python
tips.plot.scatter(x = "tip", y = "total_bill", c = tips["size"], cmap = 'coolwarm')
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001, c = tips["size"], cmap='coolwarm')
plt.colorbar()
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```

```python
# Öncelikle scatter çizerkenki amacımız; iki değişken arasındaki ilişkiyi ve dağılımı gözlemleyebilmektir.

# Seaborn ile Heatmap çizdirirken cbar=False diyerek sağdaki "size" sıkalasını kaldırabiliyoruz. 

# Ama seaborn ile scatterplot çizdirirken cbar diye bir paremetre kullanamıyoruz.

# Bunun yerine legend paremetresi ile oynayarak sağ alt tarafta gördüğünüz şekilde size sütununu boyutsal olarak renk renk gösterebiliyor.

# `legend` must be 'auto', 'brief', 'full', or a boolean.
```

```python
sns.scatterplot(x='tip', y='total_bill', hue='size', data= tips, palette='coolwarm',legend="full")
```

```python
sns.scatterplot(x='tip', y='total_bill', hue='size', data= tips, palette='coolwarm',legend= "brief")
```
