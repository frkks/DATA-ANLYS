```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

**Bar plots**

# 1. Set darkgrid style from seaborn

```python
# CODE HERE
sns.set(style="darkgrid")
```

# 2. Make a bar plot

```python
# CODE HERE
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(6, 4))
p = np.arange(len(df.index))
width = 0.20
ax.bar(p - width/2, income, width, label="income")
ax.bar(p + width/2, expense,           width, label="expense")
ax.set_xticks(p)
ax.set_xticklabels(df.index)
plt.legend()
plt.show()
```

# 3. Stack the bars

```python
# CODE HERE
total_tips = df.groupby(df.index).sum()["income"].values
total_bill = df.groupby(df.index).sum()["expense"].values
fig, ax = plt.subplots()
ax.bar(df.index,total_tips, label= "income")
ax.bar(df.index,total_bill, label = "expense")
plt.legend();
```

# 4. Rotate the labels and set figsize

```python
# CODE HERE
total_tips = df.groupby(df.index).sum()["income"].values
total_bill = df.groupby(df.index).sum()["expense"].values
fig, ax = plt.subplots(figsize=(9,6))
ax.bar(df.index,total_tips, label= "income")
ax.bar(df.index,total_bill, label = "expense")
plt.legend();
```

```python
df.plot.bar(stacked=True, rot = 90, figsize=(9,6));
```

# 5. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
# CODE HERE
index= ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
df.index = index
df
```

# 6. Unstuck the bars, annotate the hight of the bars on top of them

```python
# CODE HERE
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(9,6))
c = np.arange(len(df.index))
width = 0.20
ax.bar(c - width/2, income, width, label="income")
ax.bar(c + width/2, expense,           width, label="expense")
ax.set_xticks(c)
ax.set_xticklabels(df.index)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.001, p.get_height()+1));
plt.legend()
plt.show()
```

```python
ax = df.plot.bar(figsize = (9,6), rot=0)
ax.set_xticklabels(index)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.02, p.get_height()+0.5))
```
